## Understanding the pollutants values

Those three pollutants are harzous the humans and the ecoystems.

PM10 are small particulate matter (< 10 μm) that can be inhalate, such as solid or liquid particles contain in droplets (such dust, smoke, product of the reaction in the air of other pollutants...). Exposure to high concentrations of PM10 can result in a number of health impacts ranging from coughing and wheezing to asthma attacks and bronchitis to high blood pressure, heart attack, strokes and premature death.

When nitrogen is released during fuel combustion(traffic, factories) or from agriculture, it combines with oxygen atoms to create nitric oxide (NO). This further combines with oxygen to create nitrogen dioxide (NO2).
Nitrogen dioxide is an irritant gas, which at high concentrations causes inflammation of the airways.
NOx gases react to form smog and acid rain.


Tropospheric O3 (Ozone) refers to the O3 found at ground level. When Ozone is in the stratosphere it protect us from radiation but at ground level can damage vegetation at high concentrations.


The max anual average for NO2 stablished by the WHO and UE is 40 µg/m3, and the max per hour is 200 µg/m3.

The anual mean for PM10 20/40 µg/m3 and 50 µg/m3 per day.

Those are the limits for O3 per hour:
WHO AQG: 100 µg/m3
Valor objectiu UE:120 µg/m3

## Relationship between pollutants

```python
air_quality[["O3 µg/m3","NO2 µg/m3","PM10 µg/m3"]].corr()
```

**NO2-O3**
There is a strong relationship between the concentration of O3 and NO3. 
When nitrogen dioxide (NO2) is present in the atmosphere it breaks with the sunlight, and the O delivered react with oxigen to create tropospheric O3. That would be a possible explanation for the negative relationship (the more O3 we encounter the less NO2.

**NO2-PM10**

The realationship between NO2 and PM10 is weak and positive. 
We could think that a relationship might be found as oxides of nitrogen (NOx=NO2 + NO) are part of the reactions given in the atmosphere that results on the formation of PM.
It can be that the relationship might be another than linear or other factors are influencing.

**O3-PM10**

Those two pollutants have a weak linear relationship. 
We would expect to have some kind of realtionship because both interact in similar reaccions in the atmosphere. 
As so many other components are part of those processes maybe this is why don't see this relantionship.
