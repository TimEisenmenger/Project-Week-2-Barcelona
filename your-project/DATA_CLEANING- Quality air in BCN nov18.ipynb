# Project Week 2: The quality of air in Barcelona (November 2018)- CELANING THE DATA

## Raw data overview

```python
import pandas as pd
import numpy as np
import os
```

```python
air_quality = pd.read_csv("data/air-quality-nov-2018.csv")
```

```python
air_quality.info()
```

This data set has 5744 entries in total. As a descriptive variable we will consider Station to be used to display the missing values per column per station. We have seen already looking a bit trought de data that might a pattern of missing values for certain stations and certain pollutants.

We are gonna display a table with missing values x column and x station.

```python
#missing values in station x variable
nan_sants=air_quality[air_quality["Station"]=='Barcelona - Sants'].isnull().sum()
nan_eixample=air_quality[air_quality["Station"]=='Barcelona - Eixample'].isnull().sum()
nan_gracia=air_quality[air_quality["Station"]=='Barcelona - Gràcia'].isnull().sum()
nan_ciutadella=air_quality[air_quality["Station"]=='Barcelona - Ciutadella'].isnull().sum()
nan_valldhebron=air_quality[air_quality["Station"]=='Barcelona - Vall Hebron'].isnull().sum()
nan_palaureial=air_quality[air_quality["Station"]=='Barcelona - Palau Reial'].isnull().sum()
nan_poblenou=air_quality[air_quality["Station"]=='Barcelona - Poblenou'].isnull().sum()
nan_fabra=air_quality[air_quality["Station"]=='Barcelona - Observ Fabra'].isnull().sum()
```

```python
nan_x_station= [nan_sants,nan_eixample,nan_gracia,nan_ciutadella,nan_valldhebron,nan_palaureial,nan_poblenou,nan_fabra]

stations = list(air_quality["Station"].unique())

nan_matrix_x_column_x_station= pd.concat(nan_x_station, axis=1)

nan_matrix_x_column_x_station.columns=stations
nan_matrix_x_column_x_station = nan_matrix_x_column_x_station.drop("Station")

nan_matrix_x_column_x_station =nan_matrix_x_column_x_station.transpose()
nan_matrix_x_column_x_station
```

If we look, for instance, to Gràcia in the O3 columns, it shows 11 missing values per hour and quality and 74 for value. That makes no sense as if there is a measure of one pollutant we should be able to see the same amount of missing value for the three variables.

Looking at this on the data set, we realized that certain entries are consider as values and they should be considered as mising values.

Those are the characters considered as values and not missing values:<b>
  
**"n/a","na","--"**

## Data cleaned from special characters on NaN

### Import data and check for misssing values in one value (NaN)

This time we will import the initial dataset considering those characters as missing values.

```python
#Some of the missing values where looking like this characters
missing_values= ["n/a","na","--"]
```

```python
air_quality = pd.read_csv("data/air-quality-nov-2018.csv",na_values=missing_values)
```

```python
air_quality.head(15)
```

```python
air_quality.info()
```

```python
air_quality["Station"].unique()
```

```python
#We still have a lot of null values but we will go more in details later
air_quality.isnull().sum()
```

### Changing name of columns

```python
air_quality = air_quality.rename(columns={"O3 Value":"O3 µg/m3","NO2 Value":"NO2 µg/m3","PM10 Value":"PM10 µg/m3"})
```

```python
air_quality.head()
```

### Creating new variable Time and Date and formating it

```python
air_quality[['Generated Date','Generated Time']] = air_quality.Generated.str.split(expand=True)
air_quality
```

#### Reformat time, date variables and create a Week Day column

```python
air_quality['Generated Date'] = pd.to_datetime(air_quality['Generated Date'],format='%d/%m/%Y')
```

```python
air_quality["Week Day"] = air_quality["Generated Date"].dt.day_name()
```

```python
air_quality['Generated Date']= air_quality['Generated Date'].dt.strftime("%d/%m/%y")
```

```python
air_quality['Generated Date'] = pd.to_datetime(air_quality['Generated Date'],format='%d/%m/%y')
```

```python
air_quality.info()
```

```python
air_quality['Generated Time'] = pd.to_datetime(air_quality['Generated Time'],format="%H:%M")
```

```python
air_quality['Generated Time'] =air_quality['Generated Time'].dt.strftime("%H:%M")
```

```python
air_quality.head()
```

### Cleaning colum Station to remove Barcelona

```python
air_quality["Station"] = air_quality["Station"].str.replace('Barcelona - ', "")
```

```python
air_quality.head()
```

```python
air_quality['Station'].unique()
```

### Droping column Latitude, Longitude and Generated
