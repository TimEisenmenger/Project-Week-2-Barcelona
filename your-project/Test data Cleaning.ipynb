```python
missing_values= ["n/a","na","--"]
qair = pd.read_csv("data/air-quality-nov-2017.csv",na_values=missing_values)
```

```python
qair.head(40)
```

```python
qair = qair.rename(columns={"O3 Value":"O3 µg/m3","NO2 Value":"NO2 µg/m3","PM10 Value":"PM10 µg/m3"})
```

```python
qair[['Generated Date','Generated Time']] = qair.Generated.str.split(expand=True)
qair
```

```python
qair=qair.drop(columns =["Generated","Latitude","Longitude"])
qair
```

```python
qair["Station"] = qair["Station"].str.replace('Barcelona - ', "")
```

```python
qair[['Station']].value_counts() 
# Same number of measurement for station
```

```python
hour_columns = qair[['Station','O3 Hour','NO2 Hour','PM10 Hour','Generated Time','Generated Date','Date Time']]
```

```python
hour_columns.head(50)
#On number Date time is for a set of all stations, for the same 
#day, same hour
```

```python
#take the mean per day 03
daily_mean_03=qair.groupby(["Generated Date","Station"])[["O3 µg/m3"]].mean('O3 µg/m3')
```

```python
daily_mean_03.head(50)
```

```python
#take the mean per day no2
daily_mean_NO2=qair.groupby(["Generated Date","Station"])[["NO2 µg/m3"]].mean('NO2 µg/m3')
```

```python
daily_mean_PM=qair.groupby(["Generated Date","Station"])[["PM10 µg/m3"]].mean('PM10 µg/m3')
```
