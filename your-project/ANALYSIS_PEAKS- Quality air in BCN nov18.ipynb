```python
# Groupby to show the Air Quality acoording to pollution
air_quality.groupby(["Air Quality"]).aggregate({"O3 µg/m3":"mean","NO2 µg/m3":"mean","PM10 µg/m3":"mean"})
```

```python
air_quality_counts = air_quality.groupby(["Station"])["Air Quality"].value_counts()
air_quality_counts
```

```python
good =air_quality[air_quality['Air Quality']=="Good"][["Station"]].value_counts()
moderate =air_quality[air_quality['Air Quality']=="Moderate"][["Station"]].value_counts()
mod_good= [moderate,good]
good_mod= pd.concat(mod_good, axis=1).fillna(0)
good_mod.columns= ["Moderate", "Good" ,]

good_mod
```
