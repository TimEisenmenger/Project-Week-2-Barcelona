```python
#Displaying only the pollutants to get an overview
air_quality[["O3 Quality","O3 Value","NO2 Quality","NO2 Value","PM10 Quality","PM10 Value"]].head(9)
```

```python
#Add the unit to the columns names of the polutants
air_quality = air_quality.rename(columns={"O3 Value":"O3 µg/m3","NO2 Value":"NO2 µg/m3","PM10 Value":"PM10 µg/m3"})
air_quality
```

```python
#Split Generated into 2 columns 'Generated Date','Generated Time'
air_quality[['Generated Date','Generated Time']] = air_quality.Generated.str.split(expand=True)
air_quality
```

```python
#Remove columns "Longitude","Latitude" because we (probably) wont need them
air_quality = air_quality.drop(["Generated","Longitude","Latitude", "O3 Hour", "NO2 Hour", "PM10 Hour"], axis=1)
air_quality
```

```python
#Remove Barcelona - from the start of each object in column Station
air_quality["Station"] = air_quality["Station"].str.replace('Barcelona - ','')
air_quality
```

```python
#Transforming the column 'Generated Date' into the format of a date 
air_quality['Generated Date'] = pd.to_datetime(air_quality['Generated Date'],format='%d/%m/%Y')
air_quality["Week Day"] = air_quality["Generated Date"].dt.day_name()
air_quality['Generated Date']= air_quality['Generated Date'].dt.strftime("%d/%m/%y")
```

### Creating table of number of missing values for each column/station

```python
#missing values in station x variable
nan_sants=air_quality[air_quality["Station"]=="Sants"].isnull().sum()
nan_eixample=air_quality[air_quality["Station"]=='Eixample'].isnull().sum()
nan_gracia=air_quality[air_quality["Station"]=='Gràcia'].isnull().sum()
nan_ciutadella=air_quality[air_quality["Station"]=='Ciutadella'].isnull().sum()
nan_valldhebron=air_quality[air_quality["Station"]=='Vall Hebron'].isnull().sum()
nan_palaureial=air_quality[air_quality["Station"]=='Palau Reial'].isnull().sum()
nan_poblenou=air_quality[air_quality["Station"]=='Poblenou'].isnull().sum()
nan_fabra=air_quality[air_quality["Station"]=='Observ Fabra'].isnull().sum()
```

```python
nan_x_station= [nan_sants,nan_eixample,nan_gracia,nan_ciutadella,nan_valldhebron,nan_palaureial,nan_poblenou,nan_fabra]

stations = list(air_quality["Station"].unique())

nan_matrix_x_column_x_station= pd.concat(nan_x_station, axis=1)

nan_matrix_x_column_x_station.columns=stations

nan_matrix_x_column_x_station = nan_matrix_x_column_x_station.drop("Station")
nan_matrix_x_column_x_station =nan_matrix_x_column_x_station.transpose()

nan_matrix_x_column_x_station
```
